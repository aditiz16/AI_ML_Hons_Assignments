>**Normal Distribution**

---

A normal distribution has a probability distribution that is centered around the mean. This means that the distribution has more data around the mean. The data distribution decreases as you move away from the center. The resulting curve is symmetrical about the mean and forms a bell-shaped distribution. 

To plot the normal distribution of the dataset we use the mean and standard deviation of "Marks" column of the dataset. 

```python
marks=df["Marks"] #this will contain the list of only marks 
# calculating the mean and standard deviation  
mean=np.mean(marks)
std_dev=np.std(marks)
print("mean",mean,"Standard Deviation",std_dev)
```

From this, normal distribution data can be generated i.e to create a range of x-values and corresponding y-values based on the calculated mean and standard deviation.

Formal for y :


Note : y = _f(x)_

```python
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)
```

```python
# plotting the data using seaborn 
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# Plot the histogram
sns.histplot(marks, bins=10, color='blue',kde=True ,label='Maths Marks Distribution')
plt.title('Histogram of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))

# Plot the normal distribution curve
plt.plot(x, y, color='red', label='Normal Distribution')
plt.title('Normal Distribution of Marks')
plt.xlabel('Maths Marks')
plt.ylabel('Density')
plt.legend()
plt.show()
```

The curve is symmetric around the mean. Hence we can say that the data is evenly distributed. The standard deviation controls the spread of the curve, in this case the data spreads from -20 to +60 , it suggests variability in the data. To interpre the data more we can perform Skewness and Kurtosis

>**Measures of Skewness and Kurtosis**

---
**Skewness** is the measure of symmetry, or more precisely, the lack of symmetry. A distribution or data set is syymetric if looks the same to the left and right of the centre point. 



The skewness for a normal distribution is zero, and any symmetric data should have a skewness near zero. Negative values for the skewness indicate data that are skewed left and positive values for the skewness indicate data that are skewed right. By skewed left, we mean that the left tail is long relative to the right tail. Similarly, skewed right means that the right tail is long relative to the left tail. If the data are multi-modal, then this may affect the sign of the skewness.

Formula for Skewness :



**Kurtosis** is a measure of weather the data are heavy-tailes or light-tailed relative to a normal distribution. This, data sets with high kurtosis tend to have heavy tails, or outliers. Datasets with low kurtosis tend to have light tails or lack of outliers. A uniform distribution would be the extreme case. 

The histogram is an effective graphical technique for showing both the skewness and kurtosis of data set.



**Mesokurtic :** This distribution has a kurtosis similar to that of the normal distribution, meaning the extreme value characteristic of the distribution is similar to that of a normal distribution.

**Leptokurtic :** Any distribution that is leptokurtic displays greater kurtosis than a mesokurtic distribution. This distribution appears as a curve one with long tails (outliers.) The "skinniness" of a leptokurtic distribution is a consequence of the outliers, which stretch the horizontal axis of the histogram graph, making the bulk of the data appear in a narrow ("skinny") vertical range.

**Platykurtic :** The final type of distribution is platykurtic distribution. These types of distributions have short tails (fewer outliers.). Platykurtic distributions have demonstrated more stability than other curves because extreme price movements rarely occurred in the past

Formula for Kurtosis :







```python
# pandas has inbuilt libraries for skewness and kurtosis 

# Calculate skewness and kurtosis
skewness = marks.skew()
kurtosis = marks.kurtosis()

# Skewness plot
sns.histplot(marks, kde=True, color='blue', label='Marks Distribution')
plt.title(f'Skewness: {skewness:.2f}')
plt.xlabel('Marks')
plt.ylabel('Density')
plt.legend()
plt.show()
# Kurtosis plot
sns.histplot(marks, kde=True, color='blue', label='Marks Distribution')
plt.title(f'Kurtosis: {kurtosis:.2f}')
plt.xlabel('Marks')
plt.ylabel('Density')
plt.legend()
plt.show()
```

From the above we can see that the skewness of the Distribution is **0.66** and kurtosis is **-0.72**. It can be concluded that : 
 - The distribution is likely to have a peak that is shifted to the left, and the tail on the right side is longer than the tail on the left side.
 - The distribution's tails are lighter than those of a normal distribution, meaning it has fewer extreme values or outliers.
 - The data might be clustered more towards the lower end of the range, with a right-skewed tail extending to higher values.
