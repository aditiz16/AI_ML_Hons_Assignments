>**Correlation**

---
Correlation is a bivariate analysis that measures the strength of association between two variables and direction of the relationship. In terms of the strength of relationship the value of the correlation coefficient varies between +1 and -1. A value of +1,-1 indicates a perfect degree of association between the two variables. As the correlation coefficient values goes towards 0, the relationship is indicated by the sign of the coefficient. 
A positive correlation indicates a positive relationship whereas a negative correlation indicates a negative relationship between the variables. 


**_Pearson r correlation:_** Pearson r correlation is the most widely used correlation statistic to measure the degree of the relationship between linearly related variables.


---


>**Linear Regression T-test**

---
Linear regression is a popular statistical method used to model the relationship between a dependent variable and one or more independent variables. In linear regression, the t-test is a statistical hypothesis testing technique that is used to test the hypothesis related to linearity of the relationship between the response variable and different predictor variables.

**_T-test:_** T-tests are used in linear regression to determine if a particular variable is statistically significant in the model. A statistically significant variable is one that has a strong relationship with the dependent variable and contributes significantly to the accuracy of the model. T-tests are also used to compare the significance of different variables in the model, which can help to identify which variables are most important for predicting the dependent variable.


1. Test the association of mother's (bwt) age and birth weight using the correlation test and linear regression.

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

df=pd.read_csv('Databases/Birth.csv')

df.dtypes
```

Performing Pearson Correlation on the columns 'age' and 'bwt'

```python
corr,p=pearsonr(df['age'],df['bwt'])
print(f"Pearson Correlation Coefficient : {corr:.2f} \t p_valu : {p:.4f}")
```

By observing the pearson correlation coefficient of 0.09 and a o value of 0.2165 we can conclude that the relationship between mother's age and birth weight is very weak positively linear. This relationship is not statistically significant at the commonly used sugnificance level of 0.05.

Performing Linear Regression T-test on the columns 'age' and 'bwt' 

```python
import statsmodels.api as sm
X=df['age']
X=sm.add_constant(X)
y=df['bwt']
model=sm.OLS(y,X).fit()
summary=model.summary()
print("Summary:")
print(summary)
```

From the above we can observe that the coefficient with respect to age is 12.4297 which indicates the estimated change in 'Birth Weight' for one unit change in Mother's Age. The associated t-value is 1.240 and p value is 0.216. 
From the t-value and p value we can infer that the relationship of age of mother to that of birthweight is positive however not statistically significant at a conventional significance level of 0.05 (or 5%).
