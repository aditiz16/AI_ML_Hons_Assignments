>**Statistical Measurements**

---

**Mean (Average):** The mean is the sum of all values in a dataset divided by the number of observations. It provides a measure of the central tendency and is sensitive to outliers.
Pandas **`dataframe.mean()`** function returns the mean of the values for the requested axis. If the method is applied on a pandas series object, then the method returns a scalar value which is the mean value of all the observations in the Pandas Dataframe.

**Variance:** Variance measures the average squared deviation of each data point from the mean. A high variance indicates greater dispersion, while a low variance suggests less variability.
**`DataFrame.var(axis=None, skipna=True, ddof=1, numeric_only=False, **kwargs)`**
Return unbiased variance over requested axis.
Normalized by N-1 by default. This can be changed using the ddof argument.

**Standard Deviation:** The standard deviation is the square root of the variance. It provides a standardized measure of the spread of data and is easier to interpret than variance.
The Pandas **`Dataframe.std()`** is defined as a function for calculating the standard deviation of the given set of numbers, DataFrame, column, and rows.

**Covariance:** Covariance measures the degree to which two variables change together. Positive covariance suggests a positive relationship, while negative covariance suggests a negative relationship.

cov(x1,y1)=image.png

*  If COV(xi, xj) = 0 then variables are uncorrelated
*  If COV(xi, xj) > 0 then variables positively correlated
*  If COV(xi, xj) > < 0 then variables negatively correlated

**Correlation:** Correlation is a statistical measure that quantifies the strength and direction of a linear relationship between two variables. It indicates how changes in one variable are associated with changes in another variable. Correlation does not imply causation; it only shows that there is a consistent pattern of change between the two variables.







```python
l=df.columns.tolist()
print(*l, sep = "\n")
```

```python
l=['Sales ($billion)', 'Profits ($billion)', 'Assets ($billion)', 'Market Value ($billion)']

for i in df.columns:
  if i in l:
    mean_col = df[i].mean()
    variance_col = df[i].var()
    std_deviation_col = df[i].std()
    standard_error_col = std_deviation_col / np.sqrt(df['Company'].count())
    print("Computed Statistics for ",i)
    # Print computed statistics
    print("Mean of",i,":", mean_col)
    print("Variance of",i,":", variance_col)
    print("Standard Deviation of",i,":", std_deviation_col)
    print("Standard Error of",i,":", standard_error_col)
    print("\n")


# Calculate covariance between 'column' and 'Company'
covariance_matrix = df[l].cov()
print("Covariance Matrix:")
print(covariance_matrix)
# Calculate correlation between 'Revenue' and 'Employees'
correlation_matrix = df[l].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)






```

```python
# Display graphically the distribution of samples
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Visualization: Sales vs Profits using scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales ($billion)'], df['Profits ($billion)'], alpha=0.5)
plt.title('Sales vs Profits')
plt.xlabel('Sales ($billion)')
plt.ylabel('Profits ($billion)')
plt.show()
```

```python
# Visualization: Histogram of Asset Values
plt.figure(figsize=(10, 6))
plt.hist(df['Assets ($billion)'], bins=20, edgecolor='black')
plt.title('Distribution of Asset Values')
plt.xlabel('Assets ($billion)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization: Country Counts using a bar plot
plt.figure(figsize=(18, 6))
df['Country'].value_counts().plot(kind='bar')
plt.title('Company Counts by Country')
plt.xlabel('Country')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization: Box plot of Market Value by Continent using Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Continent', y='Market Value ($billion)')
plt.title('Market Value Distribution by Continent')
plt.xlabel('Continent')
plt.ylabel('Market Value ($billion)')
plt.xticks(rotation=45)
plt.show()
```
